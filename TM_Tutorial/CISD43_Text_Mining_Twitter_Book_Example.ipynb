**Example 1-2. Retrieving trends**

**Exploring Trending Topics**
- We are going to Twitter for the topics that are currently trending worldwide, but keep in mind that the API can easily be parameterized to constrain the topics to more specific locales if you feel inclined to try out some of the possibilities.
- The device for constraining queries is via Yahoo! GeoPlanet’s Where On Earth (WOE) ID system, which is an API unto itself that aims to provide a way to map a unique identifier to any named place on Earth (or theoretically, even in a virtual world).
- We are going to collect a set of trends for both the entire world and just the United States.

```python
# The Yahoo! Where On Earth ID for the entire world is 1.
# See https://developer.twitter.com/en/docs/trends/locations-with-trending-topics/api-reference/get-trends-available

WORLD_WOE_ID = 1
US_WOE_ID = 23424977

# Prefix ID with the underscore for query string parameterization.
# Without the underscore, the twitter package appends the ID value
# to the URL itself as a special case keyword argument.

world_trends = twitter_api.trends_place(id=WORLD_WOE_ID)
us_trends = twitter_api.trends_place(id=US_WOE_ID)

print(world_trends)
print("*******************")
print(us_trends)
```

## Extracting text, screen names, and hashtags from tweets

```python
# first let's get the text
tweets = tweepy.Cursor(twitter_api.search, q = q, lang = "en").items(count)
status_texts = [tweet.text for tweet in tweets]

# Compute a collection of all words from all tweets
words = [tweet.lower().split() for tweet in status_texts]

# Explore the first 5 items for each...
print(status_texts[:5])
print(words[:5])
```

```python
# then get the screennames
tweets = tweepy.Cursor(twitter_api.search, q = q, lang = "en").items(count)
screen_names = [tweet.user.screen_name for tweet in tweets]
print(screen_names[:5])
```

```python
# now we get the hastags
tweets = tweepy.Cursor(twitter_api.search, q = q, lang = "en").items(count)
hashtags = [tweet.entities.get('hashtags') for tweet in tweets]

# however, the amount of hashtags per tweet is different
temp = []

for hashtag in hashtags:
    # we must check if the dictionary is empty or not
    if hashtag:
        # if it isn't empty then there are hashtags
        # they are stored in a list though
        for i in hashtag:
            # below we finally grab the hashtag text
            temp.append(i["text"])
            
hashtags = temp      

print(hashtags[:5])
```

## Creating a basic frequency distribution from the words in tweets

```python
import itertools
import collections

# we must flatten the word list, but not the other two
words = list(itertools.chain(*words))

# put all the words into collections to count
common_words = collections.Counter(words)
common_screen_names = collections.Counter(screen_names)
common_hashtags = collections.Counter(hashtags)

# print out the counts
print(common_words.most_common(10))
print(common_screen_names.most_common(10))
print(common_hashtags.most_common(10))
```

## Sentiment Analysis

```python
# import
from textblob import TextBlob
import numpy as np
```

```python
# create a listener
class MyStreamListener(tweepy.StreamListener):
    # listener will include the list you want to append tweets to, and how many tweets to get
    def __init__(self, myList, limit):
        super(MyStreamListener,self).__init__()
        self.myList = myList
        self.count = 0
        self.limit = limit
    def on_status(self, status):
        # increment the count
        self.count += 1
        # if we haven't hit the limit continue streaming
        if self.count < self.limit:
            print(status.text)
            self.myList.append(status.text)
            return True
        else:
            # otherwise return false to stop streaming
            return False
    def on_error(self, status_code):
        print(status_code)
        return False
```

```python
# create a stream
tweets = []
myStreamListener = MyStreamListener(tweets, 100)
myStream = tweepy.Stream(auth = twitter_api.auth, listener=myStreamListener)
```

```python
try:
    print('Start streaming.')
    myStream.sample(languages=['en'])
finally:
    print('Done.')
    myStream.disconnect()
```

```python
# make an array to hold the scores
scores = []

# for each tweet
for i in tweets:
    # Measure the polarity of the tweet
    analysis = TextBlob(i) 
    polarity = analysis.sentiment.polarity
    
    # Store the normalized, weighted composite score
    scores.append(polarity)
```

```python
# get the most positive and negative scores
most_positive = np.argmax(scores)
most_negative = np.argmin(scores)

# test that scores was made properly
print(scores)
```

```python
# output the text of the max polarity
print('{0:6.3f} : "{1}"'.format(scores[most_positive], tweets[most_positive]))
```

```python
# output the text of the min polarity
print('{0:6.3f} : "{1}"'.format(scores[most_negative], tweets[most_negative]))
```
